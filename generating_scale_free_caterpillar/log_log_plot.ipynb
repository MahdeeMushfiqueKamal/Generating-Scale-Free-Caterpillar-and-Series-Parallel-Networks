```python
from matplotlib import pyplot as plt
import numpy as np
from util import generate_scale_free_caterpillar_graph
```

```python
graph = generate_scale_free_caterpillar_graph(100000, 0)

degree_dist = graph.degrees

plt.loglog(degree_dist.keys(), degree_dist.values())

# Label the axes
plt.xlabel("k")
plt.ylabel("N(k)")

# Show the plot
plt.show()
```

```python
import networkx as nx

graph = nx.barabasi_albert_graph(1000, 3)
```

```python
closeness_centralities = nx.closeness_centrality(graph)

# Calculate the average closeness centrality
avg_closeness_centrality = sum(closeness_centralities.values()) / len(closeness_centralities)
print("Average Closeness Centrality:", avg_closeness_centrality)
```
