## Log Log plot

```python
from util import generate_scale_free_caterpillar_graph
```

```python
graph = generate_scale_free_caterpillar_graph(100000, 0)
degree_dist = graph.degrees
```

```python
# Plot the degree distribution in log-log scale
plt.loglog(degree_dist.keys(), degree_dist.values())
sns.scatterplot(x=list(degree_dist.keys()), y=list(degree_dist.values()))

plt.ylim(1, 10000)
plt.xscale('log')
plt.yscale('log')

# Label the axes
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.title("Log-log plot of generated scale-free caterpillar network")
# Show the plot
plt.show()
```
